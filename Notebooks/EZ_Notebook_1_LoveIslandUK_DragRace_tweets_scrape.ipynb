# This notebook contains code to scrape tweets from Twitter using the `snscrape` library. It focuses on tweets containing the hashtags '#LoveIslandUK' and '#DragRace' within a specified date range. The scraped tweets are stored in pandas DataFrames for further analysis.

Scraping tweets containing the hashtag '#LoveIslandUK' within a specific date range using the `snscrape` library. The scraped tweets are stored in a pandas DataFrame.

```
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#LoveIslandUK since:2023-01-01 until:2023-06-02').get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])

    # Breaking the loop at 5000 tweets
    if i+1 == 5000:
        break

# Creating a dataframe from the tweets list above
LI_tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])

print(LI_tweets_df)
```
